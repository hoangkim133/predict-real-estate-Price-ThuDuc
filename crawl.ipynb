```python
import requests
import bs4 as bs
import pandas as pd
import numpy as np
from IPython.display import clear_output


data = pd.DataFrame(columns=['title', 'road-width', 'floors', 'bedroom', 'parking', 'dientich', 'KT', 'huong', 'dia_chi','Price'])

url_0 = 'https://alonhadat.com.vn/can-ban-nha-thanh-pho-thu-duc-ho-chi-minh-q150.htm'
url_origin = 'https://alonhadat.com.vn/can-ban-nha-thanh-pho-thu-duc-ho-chi-minh-q150/trang-'

lst_url = [url_0]
for i in range(2,410):
    s = str(i) + '.htm'
    url_final = url_origin + s
    lst_url.append(url_final)
```

```python
def checkNull(para, elem, clas):
    try:
        return para[0].find_all(elem, {"class": clas})[0].text
    except:
        return np.NaN
```

```python
for url in lst_url:
    # print(lst_url.index(url))
    # clear_output(wait = True)
    r = requests.get(url)
    convert_soup = bs.BeautifulSoup(r.text, 'lxml')
    items = convert_soup.find('div',{'class':'content-items'})
    for item in items:
        dic = {}
        if type(item) is bs.element.Tag:
            title = item.div.find_all("div", {"class": "ct_title"})
            dic['title'] = title[0].text
            text = item.find_all("div", {"class": "text"})

            # characteristics of text
            characters = text[0].find_all("div", {"class": "characteristics"})
            dic['road-width'] = checkNull(characters, "span","road-width")
            dic['floors'] = checkNull(characters, "span","floors")
            if checkNull(characters, "span","floors") == '59 lầu':
                print(url)
            bed = checkNull(characters, "span","bedroom")
            dic['bedroom'] = bed
            if (type(bed) == str):
                lst_bed = [int(s) for s in bed.split() if s.isdigit()]
                if lst_bed[0] >= 45:
                    print("bedroom=", url)

            dic['parking'] = checkNull(characters, "span","parking")

            # square-direct of text
            square = text[0].find_all("div", {"class": "square-direct"})
            dic['dientich'] = checkNull(square, "div","ct_dt")
            dic['KT'] = checkNull(square, "div","ct_kt")
            dic['huong'] = checkNull(square, "div","ct_direct")

            # price-dis of text
            price = text[0].find_all("div", {"class": "price-dis"})
            dic['dia_chi'] = checkNull(price, "div","ct_dis")
            dic['Price'] = checkNull(price, "div","ct_price")

        data = data.append(dic, ignore_index=True)




```
